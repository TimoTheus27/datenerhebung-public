Sie sind in 5 Jahresperioden beginnend ab 1950 erfasst. Ab 2020 geht es in eine Projektion über.

## Mögliche Relevante Merkmale
### CBR
**Crude birth rate (births per 1,000 population)**

Hat möglicherweise indirekten Einfluss auf die Ausbreitung von COVID-19.
### LEx
**Life expectancy at birth for both sexes combined (years)**

Hat evenutell Einfluss da eine geringe Lebenserwartung auf schlechte Lebens- und Hygienebedingungen hinweisen kann.
### GrowthRate
**Average annual rate of population change (percentage)**

Wie schnell wächst die Bevölkerung eines Landes. 
### Deaths
**Number of deaths, both sexes combined (thousands)**

Anzahl Toter in 5 Jahresperiode. Hohe Sterblichkeit kann ebenfalls auf schlechte Lebens- und Hygienebedingungen hinweisen.

### NetMigrations

Migrantenanzahl in 1000
