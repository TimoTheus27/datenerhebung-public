# Total Population Data

Dieses Set enthält Daten zur Weltbevölkerung, deren Geschlechterverteilung und Bevölkerungsdichte.

Es ist bis 2019 aktuell und geht danach in Projektionen über. 

```python
import pandas as pd

df = pd.read_csv("./data/WPP2019_TotalPopulationBySex.csv")
df.head()
```

```python
df.info()
```

## Daten von 2019
Nur die Medium Variante der Projections sichtbar.

```python
df_2019_afghanistan = df[(df["Time"] == 2019) & (df["LocID"] == 4)]

df_2019_afghanistan.head()
```

## Daten von 2020
Mehrere Varianten sichtbar, da ab hier Vorhersage.

```python
df_2020_afghanistan = df[(df["Time"] == 2020) & (df["LocID"] == 4)]

df_2020_afghanistan.head()
```
