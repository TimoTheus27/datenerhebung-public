# HDI
**Human Development Index**

Auf Deutsch "Index der menschlichen Entwicklung".

Dieser Index wird jährlich von den Vereinten Nationen veröffentlich.

Wohlstandsindikator für Staaten. 
Könnte großen Einfluss auf COVID-19 Verbreitung haben, da sich wolhabende Länder und Regionen bessere Schutzmaßnahmen leisten können. (Masken, Schutz an der Supermarktkasse, etc..)

![HDI](https://upload.wikimedia.org/wikipedia/en/2/2b/HDI_explained_the_best_way.png)

Er setzt sich aus drei Dimensionen zusammen.

$\Large HDI = \sqrt[3]{LEI \cdot EI \cdot II}$

## 1. Lebenserwartungs-Index (LEI)
$\Large LEI = \frac{LE-20}{85-20}$

LEI ist 1 wenn die Lebenserwartung bei Geburt 85 Jahre beträgt und 0 wenn diese bei 20 Jahren liegt.
## 2. Bildungs-Index (EI)
$\Large EI = \frac{MYSI + EYSI}{2}$

### 2.1 Durschnittliche-Schulbesuchdauer-Index (MYSI)
Schuljahre

$\Large MYSI = \frac{MYS}{15}$

MSY := Mean Years of Schooling 
### 2.2 Voraussichtliche-Schulbesuchdauer-Index (EYSI)
Schljahre und Universität

$\Large EYSI = \frac{EYS}{18}$

EYS := Expected Yeas of Schooling

## 3. Einkommensindex (II)
$\Large II = \frac{ln(GNIpc) - ln(100)}{ln(75000) - ln(100)}$

* II := Income Index
* GNI := Bruttonationaleinkommen(BNE)

II ist 1 wenn GNI pro Kopf 75000€ beträgt und 0 wenn das GNI pro Kopf bei 100€ liegt.
